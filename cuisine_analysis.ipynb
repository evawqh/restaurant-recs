```python
import pandas as pd
```

```python
xhs_df = pd.read_csv("data/eva-data-studio-project-01 - Xiaohongshu.csv")
to_df = pd.read_csv("data/eva-data-studio-project-01 - TimeOut.csv")
```

```python
xhs_cuisines = xhs_df.Cuisine.value_counts().to_frame()
xhs_cuisines["source"] = "Xiaohongshu"
```

```python
xhs_cuisines
```

```python
to_cuisines = to_df.Cuisine.value_counts().to_frame()
to_cuisines["source"] = "TimeOut"
```

```python
to_cuisines
```

```python
all_cuisines = pd.concat([xhs_cuisines, to_cuisines], ignore_index=False)
all_cuisines
all_cuisines = all_cuisines.reset_index(level=0)
```

```python
all_cuisines
```

```python
all_cuisines.rename({"Cuisine": "Number", "index": "Cuisine"}, axis=1, inplace=True)
```

```python
all_cuisines
```

```python
cuisines = pd.pivot_table(all_cuisines, values="Number", index="Cuisine", columns="source").fillna(0)
```

```python
cuisines
```

```python
cuisines.to_csv("all_cuisines.csv", index=True)
```
